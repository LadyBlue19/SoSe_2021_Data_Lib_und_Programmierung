```python
# ein zufälliges subset aus der gesamtmenge extrahieren
# random_state ist ein belibeiger  gewählter Startwert, den wir alle gleichsetzen um die gleiche
imdp_subsample = imdp.sample(10000, random_state=42)
```

```python
imdp_subsample.shape
```

```python
sns.stripplot(x=imdp_subsample.averageRating)
```

```python
# alpha gibt die transparente Farbe an, die zahl dahinter die Stärke der Transparenz
sns.stripplot(x=imdp_subsample.averageRating, alpha=0.1)
```

```python
sns.swarmplot(y=imdp_subsample.sample(500, random_state=1).averageRating)
```

```python
# mit der methode multiply kann eine Funktion auf ale Werte einer series angewandt werden; es wird eine neue Series zurückgegeben
# die die Rückgabewerte der Funktion enthält
def multiply_by_1000(number):
    return "blup" + str(number * 1000)

imdp_subsample.averageRating.apply(multiply_by_1000)
```

```python
# lambda-funktionen (unbennate funktionen)- wegwerf funktion an stellen, an denen eine Funktion erwahrtet wird, es aber nicht nötig ist
imdp.averageRating.apply(lambda number: "blup" + str(number * 1000))
```

```python
# wir generieren eine neue spalte (series), in der wir unsere Filme in 2 klassen einteilen
# solche die 20 (Median) oder mehr Stimmen bekommen haben und solche die weniger Stimmen bekommen haben
```

```python
imdp_subsample["voting_class"] = imdp_subsample.numVotes.apply(
    lambda number: "more_of_equal_20" if number >= 20 else "less_than_20"
)
```

```python
imdp_subsample
```

```python

sns.stripplot(data=imdp_subsample, y="averageRating", x="voting_class", alpha=0.1)
```

```python
sns.boxplot(data=imdp_subsample, y="averageRating", x="voting_class")
```
